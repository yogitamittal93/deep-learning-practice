```python
import spacy
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
```

```python
import pandas as pd
col_names = ['ID', 'Entity', 'Sentiment', 'Tweet_Content']
data = pd.read_csv('/content/twitter_training.csv', names=col_names, header=None)
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
data['sentiment_encoded'] = le.fit_transform(data['Sentiment'])


```

```python
def preprocess(text):
  doc=nlp(text)
  return " ".join([
      token.lemma_.lower()
      for token in doc
      if not token.is_stop and not token.is_punct
  ])
```

```python
nlp=spacy.load('en_core_web_sm')
vectorizer=TfidfVectorizer()
rev=[preprocess(str(text)) for text in data['Tweet_Content']]
X=vectorizer.fit_transform(rev)
y=data['sentiment_encoded']
```

```python
x_train, x_test, y_train, y_test=train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
model=SVC()
```

```python
model.fit(x_train, y_train)
```

```python
accuracy_score(y_test, model.predict(x_test))
```
